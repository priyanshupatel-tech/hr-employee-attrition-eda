# 🟢 2. Data Cleaning

```python
data.drop_duplicates(inplace=True)
```

```python
data.fillna(0,inplace=True)
```

```python
data.drop(columns=['EmployeeCount'],inplace=True)
data.drop(columns=['Over18'],inplace=True)
data.drop(columns=['StandardHours'],inplace=True)
```

# 🟢 3. Feature Engineering

### 🔹 Income Category

```python
income_category=np.where(data['MonthlyIncome']<5000,"Low Income",
                        np.where(data['MonthlyIncome']>10000,"High Income","Medium Income"))
data['income_category']=income_category
```

### 🔹 Experience Level

```python
experience_level=np.where(data['TotalWorkingYears']<5,"Fresher",
                         np.where(data['TotalWorkingYears']<9,"Mid Level","Senior"))
data['experience_level']=experience_level
```

### 🔹 Work-Life Category

```python
work_life_category=np.where(data['WorkLifeBalance']<2,"Poor",
                           np.where(data['WorkLifeBalance']<4,"Average","Good"))
data['work_life_category']=work_life_category
```

### 🔹 Attrition Risk

```python
attrition_risk=np.where((data['OverTime']=='Yes')& (data['JobSatisfaction']>2) & (data['WorkLifeBalance']<2),"High Risk",
                       np.where((data['OverTime']=='Yes') & (data['JobSatisfaction']>4) & (data['WorkLifeBalance']<4),"Medium Risk",
                               np.where((data['OverTime']=='No') & (data['JobSatisfaction']<3) & (data['WorkLifeBalance']<3),"Low risk","Medium Risk")))
data['attrition_risk']=attrition_risk
```
